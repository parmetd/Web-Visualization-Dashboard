# Initializing Exploration of Data

```python
#dependencies
import requests
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#read csv files
data_path = os.path.join("Resources/cities.csv")
cities_df = pd.read_csv(data_path)

cities_df.head()
```

```python
cities_df.describe()
```

```python
cities_df['City'].value_counts()
```

# Charts

```python
#Figure 1
plt.figure(figsize=(10,8))
plt.scatter('Lat','Max Temp', data=cities_df, color='blue')
plt.xlabel("Latitude", fontsize = 18)
plt.ylabel("r'Max Temperature ($^\circ$F)'", fontsize = 18)
plt.title("City Latitude vs. Max Temperature", fontsize = 24)
plt.ylim(-100, 150)
plt.xlim(-80, 100)
# plt.rcParams['figure.facecolor'] = 'white'
plt.grid(b=None, color='grey', which='major', axis='both', alpha=0.3)
plt.tight_layout()

# Saving visualization as an image.
plt.savefig("Charts/Figure1.png")

plt.show()
```

```python
#Figure 2
plt.figure(figsize=(10,8))
plt.scatter('Lat','Humidity', data=cities_df, color='blue')
plt.xlabel("Latitude", fontsize = 18)
plt.ylabel("Humidity (%)", fontsize = 18)
plt.title("City Latitude vs. Humidity", fontsize = 24)
plt.ylim(-20, 120)
plt.xlim(-80, 100)
plt.grid(b=None, color='grey', which='major', axis='both', alpha=0.3)
plt.tight_layout()


# Saving visualization as an image.
plt.savefig("Charts/Figure2.png")

plt.show()
```

```python
#Figure 3
plt.figure(figsize=(10,8))
plt.scatter('Lat','Cloudiness', data=cities_df, color='blue')
plt.xlabel("Latitude", fontsize = 18)
plt.ylabel("Cloudiness (%)", fontsize = 18)
plt.title("City Latitude vs. Cloudiness", fontsize = 24)
plt.ylim(-20, 120)
plt.xlim(-80, 100)
plt.grid(b=None, color='grey', which='major', axis='both', alpha=0.3)
plt.tight_layout()


# Saving visualization as an image.
plt.savefig("Charts/Figure3.png")

plt.show()
```

```python
#Figure 4
plt.figure(figsize=(10,8))
plt.scatter('Lat','Wind Speed', data=cities_df, color='blue')
plt.xlabel("Latitude", fontsize = 18)
plt.ylabel("Wind Speed (mph)", fontsize = 18)
plt.title("City Latitude vs. Windspeed", fontsize = 24)
plt.ylim(-5, 40)
plt.xlim(-80, 100)
plt.grid(b=None, color='grey', which='major', axis='both', alpha=0.3)
plt.tight_layout()

# Saving visualization as an image.
plt.savefig("Charts/Figure4.png")

plt.show()
```
